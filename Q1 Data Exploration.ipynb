## Basic Statistics

```python
# Get sorted unique advertiser id
advertiser_id = train['advertiser'].unique()
advertiser_id.sort()

# Define empty dataframes
# Train dataframe
train_df = pd.DataFrame(data=[], index=[advertiser_id], columns=['Imps', 'Clicks', 'Cost', 
                                                             'CTR', 'CPM', 'eCPC'])
# Validation dataframe
validation_df = pd.DataFrame(data=[], index=[advertiser_id], columns=['Imps', 'Clicks', 'Cost',
                                                             'CTR', 'CPM', 'eCPC'])
# Test dataframe
test_df = pd.DataFrame(data=[], index=[advertiser_id], columns=['Imps'])
```

```python
def basic_stats_calculation(dataframe,data,test=False,latex=False):
    
    advertisers = data.groupby('advertiser')
    
    # Number of impressions
    dataframe['Imps'] = advertisers['bidid'].nunique()
    
    if test is False:        
        # Number of clicks
        dataframe['Clicks'] = advertisers['click'].sum()
    
        # Total cost
        dataframe['Cost'] = advertisers['payprice'].sum()/1000
    
        # Click-through-rate
        dataframe['CTR'] = (advertisers['click'].sum() / advertisers['bidid'].nunique()) * 100

        # Cost-per-mille
        dataframe['CPM'] = dataframe['Cost'] / (advertisers['click'].count()) * 1000
                                                                    
        # Effective cost-per-click
        for i, item in enumerate(advertisers['click'].sum()):
            index = advertiser_id[i]
            if item != 0:
                dataframe.at[index,'eCPC'] = ((dataframe.at[index,'Cost']) / item)
            else:
                dataframe.at[index,'eCPC'] = 0
        dataframe['eCPC'] = dataframe['eCPC'].apply(pd.to_numeric)
        
    if latex is True:
        print(dataframe.round(3).to_latex())
        
    return dataframe
```

```python
train_df = basic_stats_calculation(train_df,train,latex=True)
```

```python
validation_df = basic_stats_calculation(validation_df,validation,latex=True)
```

```python
test_df = basic_stats_calculation(test_df,test,test=True,latex=True)
```

## eCPC
If we regard user clicks as the return, the return-on-investments (ROI) are imbalanced across the cases with different features. In order to measure the ROI effectively, this section calculates and plots the eCPC of different features.

```python
def eCPC_calculation(input_df,feature_x):
    
    group_df = input_df.groupby(feature_x)
    df = (group_df['payprice'].sum()/1000) / (group_df['click'].sum())
    for i, item in enumerate(group_df['click'].sum()):
        if item == 0:
            df[i] = 0
    df = pd.DataFrame(df.round(3),columns=['eCPC'])
    df[feature_x] = df.index
    
    return df
```

```python
def eCPC_plot(xcolumn, xlabel, ycolumn, ylabel, dataframe, modify_xvalue=False,
             sort_x=False, rotation=False):
    
    sns.set_context(rc={"lines.linewidth":0.8})
    sns.set_style('darkgrid')
    
    combine_df = pd.concat([df_1458,df_3358],axis=0)
    
    if sort_x is True:
        category = combine_df[xcolumn].unique()
        category.sort()
        sns.barplot(x=xcolumn,y=ycolumn,data=dataframe,ci=0, color='teal',
                    order=category)
    else:
        sns.barplot(x=xcolumn,y=ycolumn,data=dataframe,ci=0, color='teal')
    
    if modify_xvalue is True:
        value =combine_df[xcolumn].unique()
        value.sort()
        x_value = value+1
        ax.set_xticklabels(x_value)
    
    if rotation is True:
        plt.xticks(rotation='vertical')
    
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    return
```

```python
fig, ax = plt.subplots()
df = eCPC_calculation(df_3358,'weekday')
eCPC_plot('weekday','Weekday','eCPC', 'eCPC',df, modify_xvalue=True)
```
